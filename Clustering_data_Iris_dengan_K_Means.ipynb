```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd

#   Memuat Data ---
# Kita gunakan data Iris tetapi sengaja mengabaikan labelnya untuk clustering
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
# Label asli hanya untuk perbandingan di akhir
y_true = iris.target
```

```python
#   Inisialisasi & Latih Model K-Means ---
# Kita tentukan ingin membuat 3 cluster (karena kita tahu ada 3 spesies)
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans.fit(X)
```

```python
#   Dapatkan Hasil Cluster ---
# Model akan memberikan label cluster (0, 1, atau 2) untuk setiap data
predicted_labels = kmeans.labels_
```

```python
#  4. Visualisasi Hasil ---
plt.figure(figsize=(12, 5))
```

```python
# Plot 1: Pengelompokan oleh K-Means
plt.subplot(1, 2, 1)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=predicted_labels, cmap='viridis', edgecolor='k', s=50)
plt.title('Hasil Clustering K-Means')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
```

```python
# Plot 2: Data Asli
plt.subplot(1, 2, 2)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_true, cmap='viridis', edgecolor='k', s=50)
plt.title('Spesies Asli')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

plt.suptitle('Perbandingan Hasil Clustering dengan Data Asli', fontsize=16)
plt.show()
```
