```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

#   Memuat Data ---
# Setiap data adalah gambar 8x8 pixel, sehingga punya 64 fitur (dimensi)
digits = load_digits()
X = digits.data
y = digits.target
```

```python
#   Inisialisasi & Terapkan PCA ---
# Kita akan mereduksi 64 dimensi menjadi 2 dimensi (n_components=2)
pca = PCA(n_components=2, random_state=42)
X_reduced = pca.fit_transform(X)
```

```python
#   Visualisasi Hasil ---
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    c=y,
    cmap='viridis',
    edgecolor='k',
    alpha=0.8
)
plt.title('Dataset Digits Direduksi menjadi 2 Dimensi menggunakan PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(handles=scatter.legend_elements()[0], labels=list(range(10)), title="Digits")
plt.show()

print(f"Dimensi data asli: {X.shape}")
print(f"Dimensi data setelah reduksi: {X_reduced.shape}")
```
